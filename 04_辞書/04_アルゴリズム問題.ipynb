# 04_アルゴリズム問題2

### 🍬 アルゴリズム問題11 駄菓子屋Mix (16点)

#### 📝 問題文

- ゆるるくんは駄菓子屋へ行きました。
- 雑多に置かれた駄菓子を $ N $ 個掴みましたが、そのうちの $ S $ 番目〜 $ E $ 番目までを買うことにしました。
- 駄菓子は以下の3種類がランダムで選ばれます
  - 夢色マカロン (200円)
  - クリスタルポップ (50円)
  - 月光せんべい  (167円)
- 各駄菓子が $ D_1 $ 〜 $ D_N $ で先頭の一文字で入力される時、合計金額はいくらになるかを出力せよ

#### 📘 制約

$ 1 \leq　N \leq　10 $   
$ 1 \leq　S \leq　N $   
$ S \leq　E \leq　N $   

$ N $、$ S $、$ E $  はそれぞれ整数   
$　D_i = $　`マ` or `ク` or `月`    
$　D_i $ は1文字

#### 🔢 入力
入力は以下の形式で標準入力から与えられます

---
$ N \quad S \quad E $  
$ D_1 \quad D_2 \quad ... \quad D_N $  

---

掴んだ駄菓子の個数$ N $、買うことにした範囲（ $ S $、 $ E $）、各駄菓子 $ D_1 $ 〜 $ D_N $ を受け取ります。

#### 📤 出力

---
合計金額

---
合計金額を出力してください。

----
----

#### 🧾 入力例1
---
```python
10 1 5
マ マ マ ク ク ク ク 月 月 月
```
---

#### 🧪 出力例1
---
```python
700
```
---

※ 1番目〜5番目の駄菓子は、夢色マカロンｘ３、クリスタルポップｘ２のため、合計すると700円となります

----
----

#### 🧾 入力例2
---
```python
5 4 5
マ ク 月 月 月
```
---

#### 🧪 出力例2
---
```python
334
```
---

※ 4番目〜5番目の駄菓子は、月光せんべいｘ２のため、合計すると334円となります
