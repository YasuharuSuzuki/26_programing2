### 🎲 四択問題10 辞書内包表記 (3点)

以下のPythonコードの実行結果として正しいものはどれですか？

```python
numbers = [1, 2, 3, 4, 5]
square_dict = {n: n**2 for n in numbers if n % 2 == 0}
print(square_dict)
```

📕 1. {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}  
📗 2. {2: 4, 4: 16}  
📘 3. {1: 1, 3: 9, 5: 25}  
📙 4. {4, 16}  



````python
# 正解を出力してください
print('2')
````

---

### ✅ 解説
- 正解：📗 2. {2: 4, 4: 16}
  - `n % 2 == 0` の条件により、`numbers` の中から**偶数のみ**（2と4）が選ばれます。
  - `n: n**2` により、キーが `n`、値が `n` の2乗となるため `{2: 4, 4: 16}` が出力されます。

#### ❌ 他の選択肢が誤りである理由
- 📕 1. `if n % 2 == 0` の条件を無視して、全要素を2乗した結果であり誤りです
- 📘 3. `n % 2 == 0` は偶数を選ぶ条件ですが、これは奇数（1, 3, 5）を2乗した結果であり誤りです
- 📙 4. 辞書ではなく集合（set）のような形式になっており、キーと値のペアが失われています
