```python
# 正解を出力してください
print('2')
```

### 🎲 四択問題11 順序に基づいて渡す引数 (3点)  

- 関数呼び出し時に順序に基づいて渡す引数はのことをなんと呼ぶか？  
    📕 1. キーワード引数  
    📗 2. デフォルト引数  
    📘 3. 位置引数  
    📙 4. 可変長引数  


```python
# 正解を出力してください
print('3')
```

---

### ✅ 解説
- 正解：📘 3. 位置引数  
  - 関数呼び出し時に、**定義された順序（位置）に基づいて**渡す引数を**位置引数**と呼びます。  
  - 例えば `def greet(name, age)` に対して `greet("山田", 18)` と呼び出す場合、`"山田"` が `name`、`18` が `age` に順番通りに渡されます。  

#### ❌ 他の選択肢が誤りである理由
- 📕 1. **キーワード引数**は `greet(name="山田", age=18)` のように、引数名を指定して渡す引数です  
- 📗 2. **デフォルト引数**は `def greet(name, age=18)` のように、定義時にあらかじめデフォルト値を設定した引数です  
- 📙 4. **可変長引数**は `*args` や `**kwargs` を使って、個数が不定の引数をまとめて受け取る引数です

### 🎲 四択問題12 可変長引数 (3点)

- Pythonの関数で使われる「可変長キーワード引数」の引数定義として正しいものはどれか？  
    📕 1. param  
    📗 2. params=[]  
    📘 3. *args  
    📙 4. **kwargs  


```python
# 正解を出力してください
print('4')
```

---

### ✅ 解説
- 正解：📙 4. **kwargs
  - `**kwargs` を使うと、**個数が不定のキーワード引数**をまとめて辞書（dict）として受け取ることができます。
  - 例えば `greet(name="山田", age=18)` のように呼び出すと、`kwargs` には `{"name": "山田", "age": 18}` が格納されます。

#### ❌ 他の選択肢が誤りである理由
- 📕 1. **`param`** は特定の1つの値を受け取る通常の位置引数の定義です
- 📗 2. **`params=[]`** はデフォルト値にリストを指定したデフォルト引数の定義です
- 📘 3. **`*args`** は可変長引数ですが、キーワードなしの値をタプルとしてまとめて受け取るものであり、キーワード引数ではありません
