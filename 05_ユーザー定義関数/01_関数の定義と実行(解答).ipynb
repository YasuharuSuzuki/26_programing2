## 🔰 サンプルプログラム1　関数の定義と実行
### 🔨 関数の作り方
ユーザー定義関数は、def キーワードを使って定義され、以下のような構文を持ちます。

```python
def 関数名(引数1, 引数2, ..., 引数n):
    処理内容
    return 戻り値
```

1️⃣ `def`：関数定義の始まりを示すキーワード。  
2️⃣ 関数名：関数を識別するための名前。一般的には、関数の目的を表す英単語を使って命名します。  
3️⃣ 引数：関数に渡す入力値。0個以上の引数を持つことができます。引数は関数内で変数として使用されます。  
4️⃣ 処理内容：関数が実行するコードブロック。インデントを使って関数の範囲を示します。  
5️⃣ `return`：関数の戻り値を返すためのキーワード。戻り値がない場合、return は省略することができます。  

### 🎲 1〜6の整数をランダム返す関数
- 引数なし、戻り値1つの関数を作る

````python
# 関数の作成
# 1〜6の整数をランダムで返すシンプルなユーザー定義関数の例
from random import randint

def dice():
    result = randint(1, 6)
    return result

# 関数の使用
num = dice()
print(num)  # 出力：1〜6の整数
````

## ✌️ 練習プログラム1 (8点)
- じゃんけんの `グー`, `チョキ`, `パー` のいずれかをランダムで返す関数を作成して、呼び出してみましょう  

<!-- PROBLEM_TYPE: SYNTAX_EXERCISE -->

````python
# 1. じゃんけんの"グー","チョキ","パー"のいずれかをランダムで返す関数を作成して、呼び出してみましょう

# 正答例
import random

def janken():
    hands = ["グー", "チョキ", "パー"]
    return random.choice(hands)

# 関数の呼び出し
result = janken()
print(result)
````

---

### ✅ 正答例の解説

1️⃣ import randomでPythonの標準ライブラリであるrandomモジュールをインポートします。このモジュールは、乱数の生成に関する関数を提供します。  
2️⃣ janken()関数を定義します。この関数は引数を取らず、じゃんけんの手をランダムに返します。  
3️⃣ handsリストを定義し、じゃんけんの手である"グー", "チョキ", "パー"を要素として持つリストを作成します。  
4️⃣ random.choice(hands)を使って、handsリストからランダムに要素を選択します。random.choice()関数は、指定したリストからランダムに要素を1つ選択して返します。  
5️⃣ 選択された要素を関数の戻り値として返します。  
6️⃣ janken()関数を呼び出し、返された値をresult変数に代入します。  
7️⃣ print(result)で、選択されたじゃんけんの手を出力します。  

👉 上記のプログラムにより、実行するたびに、"グー", "チョキ", "パー"のいずれかがランダムに出力されます。
