### 𓏌𓏌 サンプルプログラム2  　2つの数値の和を返す関数
- 引数２つ、戻り値1つの関数を作る

```python
# 関数の作成
# 2つの数値の和を返すシンプルなユーザー定義関数の例
def add(a, b):
    result = a + b
    return result

# 関数の使用
added = add(3, 5)
print(added)  # 出力：8
```

1️⃣ 上記の例では、add という名前の関数を定義し、2つの引数 a と b を受け取ります。  
2️⃣ 関数は、a と b の和を計算し、その結果を result という変数に格納します。  
3️⃣ 最後に、return キーワードを使って result の値を返します。  
4️⃣ 関数を使用する際には、関数名に続いて括弧 () を記述し、必要な引数を渡します。  
5️⃣ 上記の例では、add(3, 5) のようにして関数を呼び出しています。

### ⚾️ サンプルプログラム3　マイルをメートルに変換する関数
- 引数1つ、戻り値1つの関数を作る

```python
# 関数の作成
# マイルをメートルに変換するシンプルなユーザー定義関数の例
def mile2meter(mile):
    meter = mile * 1609.344
    return meter

# 関数の使用
distance_mile = 20
distance_meter = mile2meter(distance_mile)
print(distance_meter)  # 出力：32186.88
```

1️⃣ 上記の例では、mile2meter という名前の関数を定義し、1つの引数 mile を受け取ります。  
2️⃣ 関数は、マイルからメートルを計算し、その結果を meter という変数に格納します。  
3️⃣ 最後に、return キーワードを使って meter の値を返します。  
4️⃣ 関数を使用する際には、関数名に続いて括弧 () を記述し、必要な引数を渡します。  
5️⃣ 上記の例では、mile2meter(distance_mile) のようにして関数を呼び出しています。
