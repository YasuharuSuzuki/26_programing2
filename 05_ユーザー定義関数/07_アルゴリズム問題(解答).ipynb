# 07_アルゴリズム問題

### 🔧 アルゴリズム問題13　関数版じゃんけん5回戦 (24点)

#### 📝 問題文

- AさんとBさんが５回じゃんけんをします。
- じゃんけんでグーを出したときには'G'、チョキを出したときには'C'、パーを出したときには'P'の文字がそれぞれ与えられます。
- Aさんが勝ったときにはAさんの勝利数をカウントし、Bさんが勝ったときにはBさんの勝利数をカウントします。
- あいこのときにはどちらの勝利数もカウントしません。
- ５回戦ったAさんとBさんの勝利数を出力してください。

##### ユーザー定義関数を作ること
- 今回のじゃんけん５回戦では、必ずユーザー定義関数を作って下さい
- 作成するユーザー定義関数
  - `janken` などわかり易い名前を付けましょう
  - 引数でAさんの勝ち数、Bさんの勝ち数を受け取ります
  - 関数内で input() を呼び出して、Aさん、Bさん二人の`G``C``P`を受け取りましょう
  - じゃんけんの結果によってAさんの勝ち数、Bさんの勝ち数をカウントアップします。引き分けの時はカウントアップしません
  - 戻り値でAさんの勝ち数、Bさんの勝ち数を返します
- 作成したユーザー定義関数を５回呼び出して、結果を求めて下さい
  

#### 📘 制約

- $ S_{ij} $は``G``、``C``、``P``のいずれか
- $ | S | = 1 $

#### 🔢 入力

入力は以下の形式で標準入力から与えられます

---
$ S_{A1} \quad S_{B1} $  
$ S_{A2} \quad S_{B2} $  
$ S_{A3} \quad S_{B3} $  
$ S_{A4} \quad S_{B4} $  
$ S_{A5} \quad S_{B5} $  

---

Aさんの手を$ S_{Ai}$ Bさんの手を$ S_{Bi} $で受け取ります

#### 📤 出力

 ---
$ Aさんの勝利数 \quad Bさんの勝利数 $

---
AさんとBさんの勝利数をそれぞれ出力してください。

----
----

#### 🧾 入力例1

---
```python
C P
G G
C G
P C
P P
```
---

#### 🧪 出力例1
---
```python
1 2
```
---
※ Aさん1勝、Bさん2勝のため、1 2 と出力します

----
----

#### 🧾 入力例2
---
```python
G C
C G
P P
P C
C G
```
---

### 出力例2
---
```python
1 3
```
---
※ Aさんの1勝、Bさんの3勝のため、1 3 と出力します

---
---

### ここ↓にPythonコードを記述してください

````python
# じゃんけんを１回行うユーザー定義関数を作成
def janken(a_win_count, b_win_count):
    a, b = input().split()
    ab = a + b
    if a == b:
        pass
    elif ab=='GC' or ab=='PG' or ab=='CP':
        a_win_count += 1
    else:
        b_win_count += 1
    return a_win_count, b_win_count

# ５回じゃんけんする
a_win = b_win = 0
for i in range(5):
    a_win, b_win = janken(a_win, b_win)
    
print(f"{a_win} {b_win}")
````
