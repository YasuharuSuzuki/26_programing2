### 📊 pandasについて
<img src="https://pandas.pydata.org/static/img/pandas.svg">

- pandasは、Pythonで使用できるオープンソースのデータ分析ライブラリです。その名前は「PANel DAta」の略で、パネルデータ（行と列で区切られた表形式のデータや、時系列データなど）を扱うのに優れています。
- この後の授業で機械学習を学びますが、機械学習を学ぶ時にはpandasを用いることが一般的です。

### 🎖️ pandasを使うメリット
- 便利なDataFrame、Series: DataFrame,Seriesオブジェクトは、表計算ソフトのようにデータを表形式で扱うことができる機能で、とても便利です。
- 多用なデータ形式のサポート: CSV, Excel, SQL データベース, JSON など様々な形式のデータを利用でき、それらを統一的に扱うことが可能です。
- データ分析機能: グループ化、集計、結合などの操作が簡単に行えます。また、時系列データの分析に特化した機能もあります。
- データクリーニング: 欠損値や重複データの処理が簡単にできます。また、データの型変換や正規化を容易に行うことが可能です。
- パフォーマンス: 内部の演算にはnumpyが使用されており、numpyはC言語で書かれているため、大量の数値データに対して一気に高速な計算を行うことが可能です。
- 可視化: matplotlibやseabornなどの可視化ライブラリとシームレスに連携できます。
- 機械学習ライブラリとの相性の良さ: scikit-learnなどの機械学習ライブラリとスムーズに連携できます。

## ✔　サンプルプログラム3　データの確認

#### head() / tail()
- head() --- データを先頭から何行か表示できます
- tail() --- データを終端から何行か表示できます

```python
# head() を使って先頭のデータを表示します
# 引数に何も指定しない場合、5行表示されます。
df_wine = df_wine_csv
df_wine.head()
```

```python
# 引数に行数を指定することも可能です。
df_wine.head(1)
```

```python
# tail() を使って終端のデータを表示します
df_wine.tail()
```

```python
# tail()もhead()同様に、行数を指定できます。
df_wine.head(3)
```

#### len() / shape / columns / index / info()
- len() --- 行数を取得
- shape --- 行数、列数を取得
- columns --- 列名（カラム名）を取得
- index --- インデックスを取得
- info() --- 行数、列数、データ型などを表示

```python
# 行数を取得
len(df_wine)
```

```python
# 行数、列数を取得
df_wine.shape
```

```python
# 列名（カラム名）を取得
df_wine.columns
```

```python
# インデックスを取得
df_wine.index
```

```python
# データ型の表示
df_wine.info()
```
