## 💪 難しめのプログラムを書く問題
### 🧑‍🎓 問6 (20点)
- 以下のようなネストした辞書があります。
```python
students = {
    "A組": {
        "山田太郎": {"国語": 85, "数学": 92, "英語": 78},
        "鈴木花子": {"国語": 90, "数学": 88, "英語": 95},
    },
    "B組": {
        "佐藤次郎": {"国語": 82, "数学": 90, "英語": 88},
        "田中美穂": {"国語": 95, "数学": 85, "英語": 92},
    },
}
```
- このデータは、各クラスの生徒とその生徒の教科ごとの点数を表しています。
- 全生徒の中で、各教科の最高得点を取った生徒の名前と得点を出力するプログラムを作成してください。

### 出力例
```python
国語の最高得点: 田中美穂 (95点)
数学の最高得点: 山田太郎 (92点)
英語の最高得点: 鈴木花子 (95点)
```

````python
students = {
    "A組": {
        "山田太郎": {"国語": 85, "数学": 92, "英語": 78},
        "鈴木花子": {"国語": 90, "数学": 88, "英語": 95},
    },
    "B組": {
        "佐藤次郎": {"国語": 82, "数学": 90, "英語": 88},
        "田中美穂": {"国語": 95, "数学": 85, "英語": 92},
    },
}

top_scorers = {"国語": ("", 0), "数学": ("", 0), "英語": ("", 0)}
for class_data in students.values():
    for student_name, student_scores in class_data.items():
        for subject, score in student_scores.items():
            if score > top_scorers[subject][1]:
                top_scorers[subject] = (student_name, score)

for subject, (student_name, score) in top_scorers.items():
    print(f"{subject}の最高得点: {student_name} ({score}点)")
````
