## 🔢 サンプルプログラム1　入力と出力
### 1行に1つの値
- 文字列の受け取り → 出力
- 数値の受け取り → 出力

```python
# 1行・1文字列の受け取り
text1 = input()
```

```python
print(text1) # 出力
```

```python
# 1行・1整数値の受け取り
integer1 = int(input())
```

```python
print(integer1) # 出力
```

### 🐫 1行に複数の値
- 複数の文字列を複数の変数に受け取る
- 複数の文字列を１つのリスト変数に受け取る
- 複数の整数値を複数の変数に受け取る
- 複数の整数値を１つのリスト変数に受け取る
- 複数の浮動小数値を複数の変数に受け取る
- 複数の浮動小数値を１つのリスト変数に受け取る

```python
# 複数の文字列を複数の変数に受け取る
text2, text3, text4 = input().split()
```

```python
print(text2)
```

```python
print(text3)
```

```python
print(text4)
```

```python
# 複数の文字列を１つのリスト変数に受け取る
colors1 = input().split()
```

```python
print(colors1)
```

```python
# 複数の整数値を複数の変数に受け取る
integer2, integer3, integer4, integer5 = map(int,input().split())
```

```python
print(integer2, integer3, integer4, integer5) # 出力
```

```python
# 複数の整数値を1つのリスト変数に受け取る
snacks1 = list(map(int,input().split()))
```

```python
print(snacks1)
```

```python
# 複数の浮動小数値を複数の変数に受け取る
float1, float2 = map(float,input().split())
```

```python
print(float1,float2) # 出力
```
