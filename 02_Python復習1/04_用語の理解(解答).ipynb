### 🎲 四択問題8 危険な代入 (3点)

以下のコードのうち、Pythonの予約語や組み込み関数を変数として使用しており、プログラムの動作に重大な問題を引き起こす可能性があるものはどれか？

---

#### 📕 1.
```python
print = 42
result = print + 10
```

---

#### 📗 2.
```python
number = 42
result = number + 10
print(result)
```

---

#### 📘 3.
```python
value = 42
total = value + 10
print(total)
```

---

#### 📙 4.
```python
score = 42
answer = score + 10
print(answer)
```
---

````python
# 正解を出力してください
print('1')
````

---

### ✅ 解説
- 1のコードが問題である理由
    - print はPythonの重要な組み込み関数です
    - これを変数として使用すると、以降の print() による出力が全てできなくなります
    - この状態を解消するにはPython環境の再起動が必要になります

- 2、3、4のコードは適切
    - number、value、score、result、total、answer は一般的な変数名で問題ありません
    - Pythonの予約語や組み込み関数と重複していません

- このような問題を避けるために
    - 変数名には具体的で分かりやすい一般的な名前を付ける
    - Pythonの組み込み関数の名前は変数として使用しない

### 🎲 四択問題10 リスト操作 (3点)
- リストをコロン（：）区切りで value[start:end] のように使用してリストの部分列を抽出できる機能をなんというか？   
    📕 1. ストリップ（strip）   
    📗 2. スプリット（split）    
    📘 3. スライス（slice）    
    📙 4. リスト内包表記    

````python
# 正解を出力してください
print('3')
````

---

### ✅ 解説  
- リストや文字列などのシーケンス型に対して、  
`value[start:end]` のように **コロン（:）を使って一部を取り出す操作** を **スライス（slice）** といいます。  

スライスを使うことで、  
- 指定した開始位置から終了位置まで  
- 元のデータを変更せずに  
- 新しい部分列（リスト）  
を取得できます。  

#### 🐣 **他の選択肢が誤りである理由**

**1. ストリップ（strip）**  
→ 文字列の前後にある空白や改行を削除するための機能であり、部分列の抽出ではありません。  

**2. スプリット（split）**  
→ 文字列を区切り文字で分割し、リストに変換する機能です。`start:end` 形式は使用しません。  

**4. リスト内包表記**  
→ 条件や処理を用いて新しいリストを生成する記法であり、コロンによる範囲指定とは異なります。  

👉 以上より、正解は **3. スライス（slice）** です。
