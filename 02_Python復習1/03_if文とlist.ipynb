## ❔サンプルプログラム4　if文
if文を使った条件分岐を習得しましょう
- if文
- if - else 文
- if - elif - else 文

```python
# if文
# input()で点数を受け取り、60点未満の時は”不可”と表示するプログラム
exam_score1 = int(input()) # 入力した値をintに変換して取得
if exam_score1 < 60: # 60点未満かどうかを比較。条件に適合する場合は、、、
    print("不可") # ”不可”と表示する
```

```python
# if - else 文
# input()で点数を受け取り、60点未満の時は”不可”、60点以上の場合は"可"と表示するプログラム
exam_score2 = int(input()) # 入力した値をintに変換して取得
if exam_score2 < 60: # 60点未満かどうかを比較。条件に適合する場合は、
    print("不可") # ”不可”と表示する
else: # 条件に適合しない場合は
    print("可") # ”可”と表示する
```

```python
# if - elif - else 文
# input()で点数を受け取り、90点以上の時は”秀”、80点以上の場合は"優"、70点以上の場合は"良"、60点以上の場合は"可"、
# 60点未満の場合は"不可"と表示するプログラム
exam_score3 = int(input())
if exam_score3 >= 90:
    print("秀")
elif exam_score3 >= 80:
    print("優")
elif exam_score3 >= 70:
    print("良")
elif exam_score3 >= 60:
    print("可")
else:
    print("不可")
```

## 📝 サンプルプログラム5　list
listの使い方を習得しましょう
- listを作る
- 要素の参照
- 要素の追加1 〜 append()
- 要素の追加2 〜 "+="演算子を使ったリストの結合
- 要素の削除1 〜 pop()
- 要素の削除2 〜 del
- 要素の削除3 〜 remove()
- 空のリストを作る
- range()でリストを作る
- スライス1 抽出開始位置を指定したスライス
- スライス2 抽出終了位置を指定したスライス
- スライス3 抽出開始位置、抽出終了位置を指定したスライス
- スライス4 何も指定しないスライス
- スライス5 マイナスを指定したスライス（開始位置に指定）
- スライス6 マイナスを指定したスライス（終了位置に指定）
- スライス7 マイナスを指定したスライス（開始位置と終了位置に指定）
- スライス8 stepを指定したスライス
- スライス9 stepと抽出位置を指定したスライス

```python
# listを作る
# 要素を初期化してlistを作成します
colors = ["赤","白","黄","赤","白","黄"]
print(colors)
```

```python
# 要素の参照
print(colors[3]) 
```

```python
# 要素の追加1 〜 append()で要素を追加
colors.append("紫")
print(colors)
```

```python
# 要素の追加2 〜 "+="演算子を使ったリストの結合
colors += ["黒","赤"]
print(colors)
```

```python
# 要素の削除1 〜 pop()でインデックスを指定して要素を削除。
colors.pop(2)
print(colors)
```

```python
# 要素の削除2 〜 delで範囲を指定して削除
del colors[1:4]
print(colors)
```

```python
# 要素の削除3 〜 remove()で値を指定して削除。複数ある時は前の方から削除されます
colors.remove("赤") # 黄紫黒赤
print(colors)
```

```python
# 空のリストを作る
empty_list = []
```

```python
# range()でリストを作る
range_list = list(range(5)) # 0,1,2,3,4
print(range_list)
```

```python
# スライス1 抽出開始位置を指定したスライス
print(range_list[3:])
```

```python
# スライス2 抽出終了位置を指定したスライス
print(range_list[:2])
```

```python
# スライス3 抽出開始位置、抽出終了位置を指定したスライス
print(range_list[1:3])
```

```python
# スライス4 何も指定しないスライス
print(range_list[:])
```

```python
# スライス5 マイナスを指定したスライス（開始位置に指定）
print(range_list[-3:])
```

```python
# スライス6 マイナスを指定したスライス（終了位置に指定）
print(range_list[:-2])
```

```python
# スライス7 マイナスを指定したスライス（開始位置と終了位置に指定）
print(range_list[-3:-2])
```

```python
# スライス8 stepを指定したスライス
print(range_list[::2])
```

```python
# スライス9 stepと抽出位置を指定したスライス
print(range_list[-3::2])
```

## 🪛　サンプルプログラム6　組み込み関数
min(),max(),sum()の使い方を習得しましょう
- min() を使って最小値を出力
- max() を使って最大値を出力
- sum() を使って合計値を出力

```python
# min() を使って最小値を出力
snack_price = [30, 70, 50, 90]
print(min(snack_price))
```

```python
# max() を使って最大値を出力
print(max(snack_price))
```

```python
# sum() を使って最大値を出力
print(sum(snack_price))
```
