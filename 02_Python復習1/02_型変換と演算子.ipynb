## 🎩　サンプルプログラム2　型変換
文字列を整数値に変換したり、浮動小数値を文字列に変換するなど、値の変換方法を習得しましょう
- 型の確認
- str → int
- int → str
- str → float
- float → int
- float → str

```python
# 型の確認
text5 = "1234"
integer6 = 32768
text7 = "68.47"
float3 = 170.1769246653111
print(type(text5))
print(type(integer6))
print(type(float3))
```

```python
# str → int
integer7 = int(text5)
print(integer7)
print(type(integer7))
```

```python
# int → str
text6 = str(integer6)
print(text6)
print(type(text6))
```

```python
# str → float
float4 = float(text7)
print(float4)
print(type(float4))
```

```python
# float → int
integer8 = int(float4)
print(integer8)
print(type(integer8))
```

```python
# float → str
text8 = str(float3)
text9 = "{:.2f}".format(float3)
print(text8)
print(text9)
print(type(text8))
print(type(text9))
```

## 🔢　サンプルプログラム3　算術演算子
算術演算子を使った計算を習得しましょう
- 足し算(+)
- 引き算(-)
- かけ算(*)
- 割り算(/)
- 割り算の商(//)
- 割り算の余り(%)
- べき乗(**)

```python
# 足し算(+)
num = 5 + 5
print(num)
```

```python
# 引き算(-)
num = 10 - 8
print(num)
```

```python
# かけ算(*)
num = 2 * 3
print(num)
```

```python
# 割り算(/)
num = 10 / 3
print(num)
```

```python
# 割り算の商(//)
num = 10 // 3
print(num)
```

```python
# 割り算の余り(%)
num = 10 % 3
print(num)
```

```python
# べき乗(**)
num = 2 ** 3
print(num)
```
