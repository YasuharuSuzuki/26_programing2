### 🎲 四択問題8 無名関数 (3点)

```python
squared = list(map( [ a ]  x: x**2, [1, 2, 3, 4]))
```

- 上記のコードで、空欄に入る適切なキーワードを選んでください

    📕 1. def  
    📗 2. lambda  
    📘 3. function  
    📙 4. return  


````python
# 正解を出力してください
print('2')
````

---

### ✅ 解説
- 正解：📗 2. lambda
  - `lambda` はPythonの**無名関数（ラムダ式）**を定義するキーワードです。
  - `lambda x: x**2` は「引数 `x` を受け取り、`x**2` を返す関数」を意味します。
  - `map()` の第一引数には関数を渡す必要があるため、ここでは `lambda` を使って簡潔に関数を記述しています。

#### ❌ 他の選択肢が誤りである理由
- 📕 1. **`def`** は通常の関数を定義するキーワードですが、`def` は文（statement）であるため、`map()` の引数のような式（expression）が求められる場所に直接書くことはできません
- 📘 3. **`function`** はJavaScriptなどで使われる関数定義のキーワードであり、Pythonには存在しないキーワードです
- 📙 4. **`return`** は関数の戻り値を指定するキーワードです。`lambda` 式の中では暗黙的に式の結果が返されるため、`return` を明示的に書く必要はなく、そもそも書くことができません

### 🎲 四択問題9 sort()メソッドのkey引数 (3点)

```python
numbers = [-5, 3, -1, 4, -2]
numbers.sort(key= [ a ] )
```

- 上記のプログラムで、リストの要素を絶対値の大きさでソートしたい場合、sort()メソッドのkey引数に渡す a として最も適切なものを選んでください

    📕 1. key=abs  
    📗 2. key=absolute  
    📘 3. key=lambda x: abs  
    📙 4. key="abs"  


````python
# 正解を出力してください
print('1')
````

---

### ✅ 解説
- 正解：📕 1. `abs`
  - Pythonの組み込み関数 `abs` は絶対値を返す関数です。
  - `sort()` の `key` 引数には**関数オブジェクト**を渡すことができるため、`key=abs` と書くだけで各要素に `abs()` を適用した結果を基準にソートされます。
  - 実行結果：`[-1, -2, 3, -5, 4]` → `[−1, −2, 3, 4, −5]` のように絶対値の小さい順に並び替えられます。

#### ❌ 他の選択肢が誤りである理由
- 📗 2. **`absolute`** はPythonに存在しない関数名です。絶対値を求める組み込み関数は `abs` であり、`absolute` という関数は標準では用意されていません
- 📘 3. **`lambda x: abs`** は `abs` 関数そのものを返すだけで、絶対値を**計算して返す**関数になっていません。正しくラムダ式で書くなら `lambda x: abs(x)` とする必要があります
- 📙 4. **`"abs"`** は文字列であり、関数オブジェクトではありません。`key` 引数には呼び出し可能なオブジェクト（関数）を渡す必要があるため、文字列を渡すと `TypeError` が発生します

### 🎲 四択問題10 初期化メソッド (3点)  

- Pythonでクラスの初期化メソッドを定義する際に使用する特殊メソッド名はどれですか？  
最も適切なものを選んでください。  

    📕 1. __init__()  
    📗 2. __new__()  
    📘 3. __create__()  
    📙 4. __construct__()   


````python
# 正解を出力してください
print('1')
````

---

### ✅ 解説
- 正解：📕 1. `__init__()`
  - `__init__()` はPythonのクラスにおける**初期化メソッド**（イニシャライザ）です。
  - インスタンスが生成された直後に自動的に呼び出され、インスタンス変数の設定など初期化処理を記述するために使用します。
  - 第一引数には慣習的に `self` を指定し、生成されたインスタンス自身を参照します。

```python
class Dog:
    def __init__(self, name):
        self.name = name  # インスタンス変数の初期化

dog = Dog("ポチ")
print(dog.name)  # ポチ
```

#### ❌ 他の選択肢が誤りである理由
- 📗 2. **`__new__()`** はインスタンスを**生成する**メソッドであり、`__init__()` が呼ばれる前に実行されます。通常の開発で意識する機会は少なく、初期化メソッドとは役割が異なります
- 📘 3. **`__create__()`** はPythonに存在しない特殊メソッド名です。似た名前ですが、このような標準の特殊メソッドは定義されていません
- 📙 4. **`__construct__()`** はPHPなどで使われるコンストラクタの名前であり、Pythonには存在しない特殊メソッド名です

### 🎲 四択問題11 インスタンス変数 (3点)

```python
class Animal:
    count = 0  # (1)
    
    def __init__(self, name):
        self.name = name  # (2)
        Animal.count += 1  # (3)
    
    def make_sound(self):
        sound = ""Woof!""  # (4)
        print(sound)
```

- 上記のコードで、インスタンス変数を正しく定義している部分はどれですか？

    📕 1. count = 0  # (1) の部分  
    📗 2. self.name = name  # (2) の部分  
    📘 3. Animal.count += 1  # (3) の部分   
    📙 4. sound = ""Woof!""  # (4) の部分   

````python
# 正解を出力してください
print('2')
````

---

### ✅ 解説
- 正解：📗 2. `self.name = name  # (2) の部分`
  - `self.name = name` のように **`self.` を付けて定義された変数**がインスタンス変数です。
  - インスタンス変数は各インスタンスごとに独立した値を持ち、`__init__()` 内で初期化されるのが一般的です。
  - 例えば `Animal("Dog")` と `Animal("Cat")` を作成した場合、それぞれが独立した `name` を持ちます。

#### ❌ 他の選択肢が誤りである理由
- 📕 1. **`count = 0`（(1)の部分）** はクラス直下で定義された**クラス変数**です。クラス変数はすべてのインスタンスで共有される値であり、インスタンスごとに独立した値を持つインスタンス変数とは異なります
- 📘 3. **`Animal.count += 1`（(3)の部分）** は `Animal.` を使ってクラス変数 `count` を操作している処理です。`self.` ではなくクラス名で参照しているため、これもクラス変数へのアクセスであり、インスタンス変数の定義ではありません
- 📙 4. **`sound = "Woof!"`（(4)の部分）** はメソッド内で定義された**ローカル変数**です。`self.` が付いていないため、`make_sound()` の実行中のみ存在し、インスタンスには保存されません
