## ➕ サンプルプログラム4　テキストファイルへの追記
- open()のmode引数を"a"にしてファイルを書き出すと、指定のファイルが存在しない時は新規ファイルが作られ、ファイルがある場合はその内容に追記されます。

```python
import sys
from datetime import datetime

file_path_log = "./data/log.txt"
now = str(datetime.now())
input_text = "ログ"

line = "-"*10    # 区切り線
with open(file_path_log, "a") as fileobj:    # 追記モードのファイルオブジェクト
    fileobj.write(now + "\n") 
    fileobj.write(input_text + "\n") 
    fileobj.write(line + "\n")
```
