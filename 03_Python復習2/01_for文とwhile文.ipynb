## 💫 サンプルプログラム1　for文
for文の使い方を習得しましょう
- 特定の回数繰り返す
- リストの要素を順番に取り出す
- リストの要素数だけ繰り返す
- リストの要素数だけ繰り返す2
- リスト内包表記でのfor文
- リスト内包表記を使わずに同じリストを作る

```python
# 特定の回数繰り返す
N = 5 # 繰り返したい数
for index in range(N):
    print(index)
```

```python
# リストの要素を順番に取り出す
fruits = ["apple", "banana", "orange"]
for item in fruits:
    print(item) # apple, banana, orange
```

```python
# リストの要素数だけ繰り返す
list1 = [4, 5, 2, 6, 3, 2] # 任意の配列
for index in range(len(list1)): # 配列要素数
    print(index) # 0, 1, 2, 3, 4, 5
    print(list1[index]) # 4, 5, 2, 6, 3, 2
```

```python
# リストの要素数だけ繰り返す2
list1 = [4, 5, 2, 6, 3, 2] # 任意の配列
for index, item in enumerate(list1):
    print(index) # 0, 1, 2, 3, 4, 5
    print(item) # 4, 5, 2, 6, 3, 2
```

```python
# リスト内包表記でのfor文
list1 = [1, 2, 3, 4, 5] # 任意の配列
list5 = [num1*5 for num1 in list1]
print(list5)
```

```python
# リスト内包表記を使わずに同じリストを作る
list1 = [1, 2, 3, 4, 5] # 任意の配列
list5 = []
for num1 in list1:
    list5.append(num1*5)
print(list5) # [5, 10, 15, 20, 25]
```

## 🌀 サンプルプログラム2　while文
while文の使い方を習得しましょう
- 特定の回数繰り返す
- 同じ文をfor文で書く

```python
# 特定の回数繰り返す
count = 0
while count < 4:
    print(count) # 0, 1, 2, 3
    count += 1
```

```python
# 同じ文をfor文で書く
for count in range(4):
    print(count) # 0, 1, 2, 3
```
