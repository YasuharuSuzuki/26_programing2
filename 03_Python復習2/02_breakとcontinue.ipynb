## 💔 サンプルプログラム3　break文
break文の使い方を習得しましょう
- 特定の条件でループを抜ける

```python
# 特定の条件でループを抜ける
list1 = ["a", "b", "c", "d", "e"] # 任意の配列
for index, item in enumerate(list1):
    print(item) # a, b, c, d
    if index >= 3:
        break
```

## 🔜 サンプルプログラム4　continue文
continue文の使い方を習得しましょう
- 特定の条件で次の繰り返し処理へ遷移する

```python
# 特定の条件で次の繰り返し処理へ遷移する
list1 = ["a", "b", "c", "d", "e"] # 任意の配列
for index, item in enumerate(list1):
    if (index % 2) > 0:
        continue
    print(item) # a, c, e
```
