### 🎲 四択問題5 break文と合計値 (3点)

- 以下のコードのうち、1から9までの奇数を順に処理し、5になった時点でループを終了して、それまでの合計を1度だけ出力するコードを選んでください

---

#### 📕 1.
```python
total = 0
for i in range(1, 10, 2):
    if i == 5:
        total += i
        break
print(total)
```

---

#### 📗 2.
```python
total = 0
for i in range(1, 10, 2):
    if i == 5:
        total += i
    break
print(total)
```

---

#### 📘 3.
```python
total = 0
for i in range(1, 10, 2):
    if i == 5:
        break
    total += i
    print(total)
```

---

#### 📙 4.
```python
total = 0
for i in range(1, 10, 2):
    if i == 5:
        break
    total += i
print(total)
```

````python
# 正解を出力してください
print('4')
````

---

### ✅ 解説
#### ⭕ 選択肢4が正しい理由
- i=1,3の時はtotalに加算される
- i=5の時点でループが終了する
- ループ終了後に合計値4（1+3）が出力される

#### 🔰 1,2,3が不適切な理由
1) i=1,3の時にtotalに加算せず、5の時のみ加算してからbreakするため、合計が5になる
2) 最初のループでいきなりbreak文が実行されてしまい、合計が0のまま終了する
3) 合計値が途中経過（1, 4）として出力され、最終結果のみを表示できない

### 🎲 四択問題7 例外処理 (3点)  
- 以下のプログラムで、input()で "8" と入力すると、print()で出力される文字列はどのようになりますか？

---

```python
try:
    i = int(input())
    print(8888 / i)
    print("okikokudai"[i])
except ZeroDivisionError as e:
    print('ゼロ除算が発生しました')
except IndexError as e:
    print('範囲外参照エラーが発生しました')
else:
    print('エラー発生なし！')
finally:
    print('例外処理を終了します')
print("このプログラムを正常終了します")
```

---

#### 📕 1.

```
1111.0
ゼロ除算が発生しました
例外処理を終了します
このプログラムを正常終了します
```
---

#### 📗 2.

```
1111.0
範囲外参照エラーが発生しました
例外処理を終了します
このプログラムを正常終了します
```
---

#### 📘 3.

```
1111.0
a
エラー発生なし！
例外処理を終了します
このプログラムを正常終了します
```
---

#### 📙 4.

```
例外処理を終了します
```


````python
# 正解を出力してください
print('3')
````

---

### ✅ 解説

- `input()` に `"8"` を入力すると、`i = 8` となる。

- `8888 / 8` はゼロ除算ではないため
→ `1111.0` が出力される

- `"okikokudai"[8]` は文字列の範囲内で
→ `'a'` が出力される

- `try` 内で例外は発生しないため
→ `else` が実行され「エラー発生なし！」が出力される

- `finally` は必ず実行される

- `try-except` の後の `print()` も実行される

👉 以上より、正解は **3.** です。

### 🎲 四択問題8 set (3点)

以下のプログラムを実行します

```python
all_color = set(["赤","白","黄","紫","桃"])
input_color = set({"赤","白","白","赤","白","黒"})
print(len(all_color-input_color))
```

- 出力結果を下記から選べ

    📕 1. {'紫', '黄'}   
    📗 2. {'桃', '黒, '黄'}   
    📘 3. 2   
    📙 4. 3   


````python
# 正解を出力してください
print('4')
````

---

### ✅ 解説  

`set` は 重複を除外します  
→ `input_color = {'赤', '白', '黒'}`  

`all_color - input_color` によって  
all_color にあって input_color にない要素が残ります  
→ `{'黄', '紫', '桃'}`

`len()` でsetの要素数を出力します  
→ 3個  

👉 以上より、正解は **4.** です。
